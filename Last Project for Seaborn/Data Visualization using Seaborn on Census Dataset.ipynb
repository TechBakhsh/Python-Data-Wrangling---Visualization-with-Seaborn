# Import Main Library

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("census_data.csv",names=["age", "work", "work_ID", "Education", "Education-num", "marital-status", 
           "Occupation", "Relationship", "Race", "Gender", "Capital-gain", "Capital-loss",
           "Hours-per-week", "Native-country", "salary"])
```

# Data Visualization

```python
# Assuming df is your DataFrame
sns.scatterplot(x="Capital-loss", y="Capital-gain", data=df, hue="marital-status")

# Add labels and title
plt.xlabel("Capital Loss")
plt.ylabel("Capital Gain")
plt.title("Scatter Plot of Capital Loss vs Capital Gain with Marital Status")

# Show the plot
plt.show()
```

```python
# Assuming df is your DataFrame
sns.scatterplot(x="age", y="Capital-gain", data=df, hue="marital-status")

# Add labels and title
plt.xlabel("age")
plt.ylabel("Capital Gain")
plt.title("Scatter Plot of Capital Loss vs Capital Gain with Marital Status")

# Show the plot
plt.show()
```

```python
# Assuming df is your DataFrame
sns.relplot(x="age", y="Capital-gain", data=df, hue="marital-status")

# Add labels and title
plt.xlabel("age")
plt.ylabel("Capital Gain")
plt.title("Scatter Plot of Capital Loss vs Capital Gain with Marital Status")

# Show the plot
plt.show()
```

```python
# Assuming df is your DataFrame
sns.scatterplot(x="Education-num", y="Capital-gain", data=df, hue="age")

# Add labels and title
plt.xlabel("Education-num")
plt.ylabel("Capital Gain")
plt.title("Scatter Plot of Capital Loss vs Capital Gain with Marital Status")

# Show the plot
plt.show()
```

```python
sns.relplot(x="Education-num", y="Capital-gain", data=df, hue="age", col='Gender', size="age", kind="scatter")

# Add labels and title
plt.xlabel("Education-num")
plt.ylabel("Capital Gain")
plt.suptitle("Scatter Plot of Education-num vs Capital Gain with Age and Gender", y=1.02)

# Show the plot
plt.show()
```

# Line plot

# Categorical part

```python
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

sns.catplot(x="age", y="Education", data=df, kind="box", height=6, aspect=2, hue="Gender")

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Education")
plt.title("Categorical Box Plot of Age vs Education with Gender Hue and Jitter")

# Show the plot
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

sns.violinplot(x="age", y="marital-status", data=df, hue="Gender", inner="quartile", split=True, palette="muted", jitter=True)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("marital-status")
plt.title("Violin Plot of Age vs Education with Gender Hue and Jitter")

# Show the plot
plt.show()
```

# Bar Plot


```python
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

sns.barplot(x="age", y="Gender", data=df, hue="Gender", palette="muted", ci="sd", dodge=True)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Gender")
plt.title("Bar Plot of Age vs Education with Gender Hue")

# Show the plot
plt.show()
```

# Point Plot

```python
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

sns.pointplot(x="age", y="Education", data=df, hue="Gender", markers="o", linestyles="--", dodge=True)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Education")
plt.title("Point Plot of Age vs Education with Gender Hue")

# Show the plot
plt.show()
```

# Heat Map

```python
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed

# Pivot the DataFrame to create a matrix suitable for a heatmap
heatmap_data = df.pivot_table(index="Education", columns="age", aggfunc="size")

# Create the heatmap
sns.heatmap(heatmap_data, cmap="viridis", annot=True, fmt=".2f", linewidths=.5)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Education")
plt.title("Heatmap of Age vs Education")

# Show the plot
plt.show()
```

# Pair Plot

```python
# Example with specific variables
sns.pairplot(df, vars=["age", "Education-num", "Capital-gain", "Capital-loss"])

# Show the plot
plt.show()
```
