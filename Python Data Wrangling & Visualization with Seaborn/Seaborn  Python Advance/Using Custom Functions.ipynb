```python
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Create a Q-Q plot for "total_bill" using statsmodels
fig, ax = plt.subplots(figsize=(6, 6))
sm.qqplot(tips["total_bill"], line="45", ax=ax)
ax.set_title("Q-Q Plot for Total Bill")
plt.show()

# Create a PairGrid with scatter plots for "total_bill", "tip", and "size"
g = sns.PairGrid(tips, vars=["total_bill", "tip", "size"])
g.map_upper(sns.scatterplot, alpha=0.7)
g.map_diag(sns.histplot)
g.add_legend()
plt.suptitle("PairGrid: Scatter Plots for Total Bill, Tip, and Size", y=1.02)
plt.show()
```
