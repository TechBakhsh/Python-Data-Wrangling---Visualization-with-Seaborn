```python
# Import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks")
tips=sns.load_dataset("tips")
g=sns.FacetGrid(tips,col="time")
plt.show()
```

```python
g=sns.FacetGrid(tips,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
g=sns.FacetGrid(tips,col="time", hue="smoker")
g.map(plt.scatter,"total_bill","tip",alpha=.7)
plt.show()
```
