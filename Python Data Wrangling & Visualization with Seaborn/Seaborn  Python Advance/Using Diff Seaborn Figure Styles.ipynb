```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# List of Seaborn styles to try
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

# Create subplots for each style
fig, axes = plt.subplots(nrows=1, ncols=len(styles), figsize=(15, 4))

# Generate x values
x = np.linspace(0, 14, 100)

for style, ax in zip(styles, axes):
    # Set the style
    sns.set(style=style)
    
    # Create a sine wave plot
    sns.lineplot(x=x, y=np.sin(x), ax=ax)
    
    # Set a title
    ax.set_title(f'Style: {style.capitalize()}')

# Adjust layout for better appearance
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plot
plt.show()
```
