```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Create a pairplot for the entire dataset
sns.pairplot(tips, diag_kind='hist', markers='o', hue='sex', palette='viridis')

# Add a title
plt.suptitle("Pairwise Data Relationships for the 'tips' Dataset", y=1.02)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a pairplot for the entire dataset
sns.pairplot(iris, hue='species', markers='o', palette='viridis')

# Add a title
plt.suptitle("Pairwise Data Relationships for the 'iris' Dataset", y=1.02)

# Show the plot
plt.show()
```

```python
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

# Create a sample dataset (replace this with your own dataset)
data = {
    'feature1': [1, 2, 3, 4],
    'feature2': [5, 6, 7, 8],
    'feature3': [9, 10, 11, 12],
}

df = pd.DataFrame(data)

# Create a scatter plot matrix for pairwise relationships
scatter_matrix(df, alpha=0.8, marker='o', c='red', figsize=(8, 8), diagonal='hist')

# Add a title
plt.suptitle("Pairwise Scatter Plots for Custom Dataset", y=1.02)

# Show the plot
plt.show()
```
