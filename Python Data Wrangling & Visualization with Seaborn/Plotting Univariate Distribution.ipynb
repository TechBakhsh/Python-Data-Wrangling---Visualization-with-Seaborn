```python
# import library
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
# Set color codes
sns.set(color_codes=True)

# Generate random data from a normal distribution
x = np.random.normal(size=100)

# Plot the distribution
sns.distplot(x)

# Show the plot
plt.show()
```

```python
sns.distplot(x,bins=20,kde=False,rug="True")
plt.show()
```

```python
sns.distplot(x,bins=20,hist=False,rug="True")
plt.show()
```

```python
# Load a dataset (you can replace this with your own data)
tips = sns.load_dataset("tips")

# Plot the univariate distribution using a histogram
sns.histplot(tips["total_bill"], kde=True, bins=20, color='skyblue')

# Add labels and title
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.title("Univariate Distribution of Total Bill Amount")

# Show the plot
plt.show()
```

```python
tips = sns.load_dataset("tips")

# Plot the univariate distribution using a rug plot only
sns.rugplot(tips["total_bill"], height=5.8, color='green', alpha=0.5)  # Adjust height and color as needed

# Add labels and title
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.title("Univariate Distribution of Total Bill Amount (Rug Plot Only)")

# Show the plot
plt.show()
```

```python
sns.kdeplot(tips["total_bill"], color='skyblue', linestyle='--', linewidth=2)
plt.show()
```
