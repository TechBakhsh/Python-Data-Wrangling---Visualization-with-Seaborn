```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Create a scatter plot with a linear regression line for the relationship between "total_bill" and "tip"
sns.regplot(x="total_bill", y="tip", data=tips, order=1, line_kws={"color": "red"})

# Add labels and title
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Linear Regression: Total Bill vs Tip")

# Show the plot
plt.show()
```
