```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Create a joint plot for "total_bill" and "tip"
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")

# Show the plot
plt.show()
```

```python
# Create a joint plot with a regression fit
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")

# Show the plot
plt.show()
```

```python
# Create a bivariate KDE plot for "total_bill" and "tip"
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde", fill=True, cmap="Blues")

# Show the plot
plt.show()
```

```python
# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Create a PairGrid with bivariate KDE plots
g = sns.PairGrid(tips)
g.map_upper(sns.kdeplot, cmap="Blues", fill=True)
g.map_lower(sns.scatterplot)
g.map_diag(sns.histplot, kde_kws={"color": "k"})

# Show the plot
plt.show()
```
