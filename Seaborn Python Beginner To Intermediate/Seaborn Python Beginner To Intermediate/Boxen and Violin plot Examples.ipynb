```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
tips=sns.load_dataset("tips")
ax=sns.boxenplot(x=tips["total_bill"])
plt.show()
```

```python
sns.boxenplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.boxenplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

```python
sns.boxenplot(x="day", y="total_bill", hue="time", data=tips,linewidth=2.5, palette="Set3")
plt.show()
```

```python
g=sns.catplot(x="sex", y="total_bill", hue="smoker",data=tips,col="time" , height=4,kind="boxen",)
plt.show()
```

```python
tips = sns.load_dataset("tips")
ax = sns.violinplot(x=tips["total_bill"])
plt.show()
```

```python
tips = sns.load_dataset("tips")
ax = sns.violinplot(x="day", y="total_bill",data=tips)
plt.show()
```

```python
tips = sns.load_dataset("tips")
ax = sns.violinplot(x="day", y="total_bill",data=tips,hue="smoker",palette="muted")
plt.show()
```

```python
tips = sns.load_dataset("tips")
ax = sns.violinplot(x="day", y="total_bill",data=tips,split=True,hue="smoker",palette="muted")
plt.show()
```

```python
tips = sns.load_dataset("tips")
ax = sns.violinplot(x="time", y="total_bill",data=tips, order=["Lunch","Dinner"],palette="muted")
plt.show()
```

```python
tips = sns.load_dataset("tips")
ax = sns.violinplot(x="day", y="total_bill",data=tips,split=True,scale="count",palette="muted", inner="quartile")
plt.show()
```

```python
tips = sns.load_dataset("tips")
ax = sns.violinplot(x="day", y="total_bill",data=tips,split=True,scale="count",palette="muted", inner="stick")
plt.show()
```
