```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# load data
sns.set_style("darkgrid")
tips= sns.load_dataset("tips")
# create pointplot
sns.pointplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# create pointplot
sns.pointplot(x="day", y="total_bill",hue="smoker",data=tips)
plt.show()
```

```python
# Create a categorical swarm plot without connecting the points
sns.pointplot(x="tip", y="day", data=tips, )

# Show the plot
plt.show()
```

```python
# Create a categorical swarm plot without connecting the points
sns.pointplot(x="tip", y="day", data=tips, join=False)

# Show the plot
plt.show()
```

```python
sns.pointplot(x="time",y ="total_bill", hue="smoker",data=tips,palette="Set2")
plt.show()
```

```python
sns.pointplot(x="time",y="tip",order=["Lunch", "Dinner"], data=tips)
plt.show()
```

```python
sns.pointplot(x="day", y="tip",estimator="median", data=tips)
plt.show()
```

```python
sns.pointplot(x="day", y="tip",estimator="median", ci=68, data=tips)
plt.show()
```

```python
sns.pointplot(x="day", y="tip",estimator="median", data=tips,cl="sd")
plt.show()
```

```python
# Create a point plot with facets
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="point", dodge=True, height=4, aspect=.4)

# Show the plot
plt.show()
```
