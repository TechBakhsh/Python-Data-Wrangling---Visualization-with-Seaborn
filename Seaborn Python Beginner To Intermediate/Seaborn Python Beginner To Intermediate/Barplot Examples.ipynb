```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# load data
tips=sns.load_dataset("tips")
ax=sns.barplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
tips=sns.load_dataset("tips")
ax=sns.barplot(x="time",y="total_bill", order=["Lunch", "Dinner"],data=tips)
plt.show()
```

```python
tips=sns.load_dataset("tips")
ax=sns.barplot(x="day",y="total_bill", estimator="median",data=tips)
plt.show()
```

```python
tips=sns.load_dataset("tips")
ax=sns.barplot(x="day",y="total_bill",ci=68, estimator="median",data=tips)
plt.show()
```

```python
tips=sns.load_dataset("tips")
ax=sns.barplot(x="day",y="total_bill",ci="sd", estimator="median",data=tips)
plt.show()
```

```python
tips=sns.load_dataset("tips")
ax=sns.barplot(x="day",y="total_bill",ci="sd",capsize=.2, estimator="median",data=tips)
plt.show()
```

```python
ax = sns.barplot(x="day", y="total_bill", ci="sd", palette="Blues", estimator="median", data=tips)
plt.show()
```

```python
ax = sns.barplot(x="day", y="total_bill", saturation=.5, color="salmon", data=tips)
plt.show()
```

```python
g= sns.catplot(x="sex", y="total_bill", hue="smoker", height=4, kind="bar", col="time", data=tips)
plt.show()
```
