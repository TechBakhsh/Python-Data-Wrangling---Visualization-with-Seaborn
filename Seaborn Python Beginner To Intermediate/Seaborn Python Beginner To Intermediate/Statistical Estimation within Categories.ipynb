```python
# import library
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived",hue="class", kind="bar", data=titanic)
plt.show()
```

```python
# Create a count plot
sns.catplot(x="deck", kind="count", palette="ch:.5", data=titanic)

# Show the plot
plt.show()
```

```python
# Create a categorical count plot with specified edgecolor
sns.catplot(
    y="sex",
    hue="class",
    kind="count",
    palette="pastel",
    edgecolor="black",  # Set the edgecolor to black
    data=titanic
)

# Show the 
```

```python
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
plt.show()
```

```python
# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a categorical strip plot
sns.catplot(x="species", y="sepal_length", kind="strip", data=iris)

# Show the plot
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a categorical swarm plot
sns.catplot(x="day", y="total_bill", col="time", aspect=0.6, hue="smoker", kind="swarm", data=tips)

# Show the plot
plt.show()
```
