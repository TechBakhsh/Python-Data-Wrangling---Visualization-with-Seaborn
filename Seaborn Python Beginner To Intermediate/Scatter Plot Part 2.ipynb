```python
#Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(style="darkgrid")
tips=sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);
plt.show()
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", palette=cmap, legend="full", data=tips)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")  # Example dataset, replace with your actual data
frequency, wide = 30, 3  # Replace with your actual values

# Create scatterplot
sns.scatterplot(x="total_bill", y="tip", hue="day", size="smoker", palette="Set2", data=tips)

# Add specific points
points = [(1, 1, 'a'), (1, 2000, 'b'), (frequency, wide, 'c')]

for x, y, label in points:
    plt.scatter(x, y, color='red', marker='*', s=100)  # Customize the appearance of the point
    plt.text(x, y, label, fontsize=12, ha='right', va='bottom')  # Add labels

# Show the plot
plt.show()
```
