```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Example 1: Line plot with fmri dataset and facets
sns.relplot(x="timepoint", y="signal", hue="subject", col="region", row="event", height=3, kind="line", estimator=None, data=fmri)

# Automatically format the x-axis as dates
plt.show()

# Example 2: Line plot with random data and facets
df = pd.DataFrame(dict(time=pd.date_range("2002-1-1", periods=500), value=(np.random.randn(500).cumsum())))
g = sns.relplot(x="time", y="value", kind="line", data=df)

# Automatically format the x-axis as dates
g.fig.autofmt_xdate()

# Example 3: Line plot with fmri dataset without facets
palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(x="timepoint", y="signal", hue="event", style="event", kind="line", data=fmri)

# Show the plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot with facets and query for the 'frontal' region
sns.relplot(
    x="timepoint",
    y="signal",
    hue="event",
    style="event",
    col="subject",
    col_wrap=5,
    height=3,
    aspect=.75,
    linewidth=2.5,  # Corrected syntax
    kind="line",
    data=fmri.query("region == 'frontal'")  # Corrected syntax
)

# Show the plot
plt.show()
```
